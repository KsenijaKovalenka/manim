```python
class Qbit(VGroup):
    def __init__(self, radius=2):
        super().__init__()
        c = Circle(radius=radius)
        lx = Arrow(ORIGIN, (DL * (radius+1)) * 0.7, color=TEAL, buff=0)
        ly = Arrow(ORIGIN, RIGHT * (radius+1), color=TEAL, buff=0)
        lz = Arrow(ORIGIN, UP * (radius+1), color=TEAL, buff=0)
        
        #arrow = Arrow(ORIGIN, UP * radius, buff=0, color=PURPLE)

        x_label = MathTex("x").next_to(lx, DL)
        y_label = MathTex("y").next_to(ly, RIGHT)
        z_label = MathTex("z").next_to(lz, UP)

        self.add(
          c, lx, ly, lz,
          #arrow,
          x_label, y_label, z_label
        )
```

```python
# for 1
class Example(Scene):
    def construct(self):
        number_plane = NumberPlane().set_opacity(0.5)

        qbit1 = Qbit().to_edge(LEFT)
        qbit2 = Qbit().to_edge(RIGHT)
        arrow1 = Arrow(ORIGIN, UP * 2, buff=0, color=PURPLE).move_to(UP + 4*LEFT)
        arrow2 = Arrow(ORIGIN, UP * 2, buff=0, color=PURPLE).move_to(UP + 3.13*RIGHT)
        self.add(
            number_plane,
            qbit1, qbit2
        )
        self.play(
            Rotate(arrow1, (-1)*-2.91, about_point=arrow1.get_start()),
            Rotate(arrow2, (-1)*-1.32, about_point=arrow2.get_start()),
            run_time=2
        )
        self.wait()
        self.play(
            Rotate(arrow1, (-1)*6.07, about_point=arrow1.get_start()),
            Rotate(arrow2, (-1)*4.39, about_point=arrow2.get_start()),
            run_time=2
        )
        self.wait()
        


%manim $_RV
```
